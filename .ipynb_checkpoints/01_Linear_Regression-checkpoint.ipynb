**Hypothesis (Best Fit Line):**

$h_{\theta}(x) = \theta_0 + \theta_1 \cdot x$

**Cost Function:**

$J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} \left[ h_{\theta}(x_i) - y_i \right]^2$


```python
# Code to visualize MSE (Mean square error) cost function in 3D space. Generated by chat gpt with following prompt : 
''' 
Prompt 1 : 
I am studying the concepts of machine learning. I have started simple linear regression. 

Lets say the best fit line that I am trying to find is : 
h_theeta(x) = theeta_0 + theeta_1*x 

and cost function is : 
J(theeta0,theeta1) = (1/2m)* sum(i=1 to m) [  ( h_theeta(x_i) - (y_i)  )^2 ] 
Please list down all the above formula in jupyter notebook markdown format 

Check response ... and confirm if it is ok

Prompt 2 : 
Write a code in python to : 
1. generate sample data for x,y
2. generate sample values for theeta_0 , theeta_1 .. and calculate : 
2.1 : h_theeta(x) = theeta_0 + theeta_1*x .. for each combination of theeta_0, theeta_1 generated 
2.2 :  J(theeta_0, theeta_1) for each combination of theeta_0, theeta_1 generated ... using the cost function formula shared in the previous response.
3. plot a 2d scatter plot - showing all (x,y) points ...
4. Adjacent to the plot in step 3, plot a 3d curve plot showing theeta_0, theeta_1 in x,y axis .. and J(theeta_0,theeta_1) in z axis

'''

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 1. Generate sample data for x, y
np.random.seed(42)  # For reproducibility
m = 1000  # Number of data points
x = np.random.rand(m, 1) * 10  # Generate x values between 0 and 10
theta_actual = [2, 3]  # Actual theta_0 and theta_1 for generating y
y = theta_actual[0] + theta_actual[1] * x + np.random.randn(m, 1)  # Generate y values

# 2. Generate sample values for theta_0, theta_1
theta_0_vals = np.linspace(-1500, 1500, 100)
theta_1_vals = np.linspace(-1500, 1500, 100)
theta_0, theta_1 = np.meshgrid(theta_0_vals, theta_1_vals)

# 2.1 Calculate h_theta(x) for each combination of theta_0, theta_1
h_theta = np.zeros((len(theta_0_vals), len(theta_1_vals), m))
for i in range(len(theta_0_vals)):
    for j in range(len(theta_1_vals)):
        h_theta[i, j] = theta_0[i, j] + theta_1[i, j] * x.flatten()

# 2.2 Calculate J(theta_0, theta_1) for each combination using the cost function
J_theta = np.zeros(theta_0.shape)
for i in range(len(theta_0_vals)):
    for j in range(len(theta_1_vals)):
        J_theta[i, j] = (1 / (2 * m)) * np.sum((h_theta[i, j] - y.flatten()) ** 2)

# 3. Plot a 2D scatter plot showing all (x, y) points
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', label='Data Points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D Scatter Plot of Data Points')
plt.legend()

# 4. Plot a 3D surface plot of theta_0, theta_1 vs J(theta_0, theta_1)
ax = plt.subplot(1, 2, 2, projection='3d')
ax.plot_surface(theta_0, theta_1, J_theta, cmap='viridis', alpha=0.8)
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
ax.set_zlabel(r'$J(\theta_0, \theta_1)$')
ax.set_title('3D Surface Plot of Cost Function')

plt.tight_layout()
plt.show()
```

### Adjusted R2

Adjusted R2 = 1 - \[ (1-R2) * (n-1) / (n-p-1) \] 
<br>n = number of rows
<br>p = number of features
<br>
<br>Adjusted R2 cannot be calculated using sklearn .. as it requires information on number of features and number of rows. It is calculated using formula

```python
## R-sqr and adj R-sqr
from sklearn.metrics import r2_score

r_sqr = r2_score(y_test,y_test_pred)
adj_r_sqr = 1-(1-r_sqr)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)

print(r_sqr)
print(adj_r_sqr)
```

### K fold Cross Validation
- The code for cross-validation is important
- In CV, Train data is split into train and validation
- If Train datasize is 900 rows and CV=5, then 900/5 = 180 ... the algo will split modeling into the following experiments :
    - Experiment 1 : First 180 rows test ... remaining train .... sklearn cross validation then trains data on this subset .. and performs validation on the test set ... 
    - Experiment 2 : Second 180 rows test .. remaining train
    - and so on
- For every experiment, The specified performance metrics will be calculated on each of the 5 test subsets ... and we will get 5 performance metric as the output
- The average of the scores can be calculated to find the metric for the entire training data - this is the accepted practice.
- The goal of cross validation is to see if the model is able to be trained consistently across all the k folds ... and then the final model is still trained on the whole training data .... k fold is just a validation technique to find out the training performance metric

```python
from sklearn.model_selection import cross_val_score
validation_score = cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=3)
```

## Subsetion of 36.2 - Pipeline in polynomial

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    X_new = np.linspace(-3, 3, 200).reshape(200, 1)
    
    poly_features=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression()
    poly_regression=Pipeline([
        ("poly_features",poly_features),
        ("lin_reg",lin_reg)
    ])
    poly_regression.fit(X_train,y_train) ## Because we are using pipeline() - this will call fit of ploynomial first and then fit of linear regression
    y_pred_new=poly_regression.predict(X_new) ## Because we are using pipeline() - this will call predict of ploynomial and then predict of linear regression
    
    ## Suvam not sure how, but in pipeline .fit calls .fit_transform automatically for the polynomial object - maybe python handles it internally .
    ## Similarly .predict calls .transform automatically for the polynomial object
    ## Need more clarity on Pipeline - its unclear how it works

    
    #plotting prediction line
    plt.plot(X_new, y_pred_new,'r', label="Degree " + str(degree), linewidth=2)
    plt.plot(X_train, y_train, "b.", linewidth=3)
    plt.plot(X_test, y_test, "g.", linewidth=3)
    plt.legend(loc="upper left")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4, 0, 10])
    plt.show()
    
```

## Parabola Basics by Suvam

1. Parabola is always symmetric around its vertex ... along some axis
2. Multiplying a number > 1 with x^2 results in steeper parabola .. Illustration below

```python
# Dont review code - just look at output
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-10, 10, 400)

# Define the functions with a larger coefficient for the linear term
y_parabola = 2 * x**2         
y_line = 4 * x**2                
#y_combined = y_parabola + 10 * x   

# Plotting
plt.figure(figsize=(6, 6))

# Plot the parabola
plt.plot(x, y_parabola, label=r'$y = 2x^2$', color='blue')

# Plot the line
plt.plot(x, y_line, label=r'$y = 4x^2$', color='green')

# Plot the combined function
#plt.plot(x, y_combined, label=r'$y = 2x^2 + 10x$', color='red')

# Adding labels and title
plt.xlabel('x -->')
plt.ylabel('y -->')


# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

3. y = (x-2)^2 + 4 --> Shifts the vertex of the parabola to (2,4) - in the same shape as before. Illustration below. Notice that the steepness of the curve is the same - as the same factor is multiplied to x^2 as in the original curve... and the curve is symmetric around the vertex.. Just the vertex has shifted.
4. Any equation y = ax^2 + bx + c .. can be represented as y = a(x-k)^2 + c ... i.e. it will be a parabola with vertex shifted to a different point .. and some factor multiplied - which increases the steepness of the curve.  
5. If parabola is shifted, and the same factor is multipled to x^2 as in the original curve, then the curves will intesect at exactly one point. The same illustration below captures this as well. 

```python
# Dont review code - just look at output
import numpy as np
import matplotlib.pyplot as plt

# Define the range of x values
x = np.linspace(-5, 5, 400)

# Define the functions with a larger coefficient for the linear term
y_parabola = x**2         

# Plot the parabola
plt.plot(x, y_parabola, label=r'$y = x^2$', color='blue')
plt.xlabel('x -->')
plt.ylabel('y -->')
plt.grid(True)
plt.show()

x = np.linspace(-3, 7, 400)
y_line = (x-2)**2 + 4                
#y_combined = y_parabola + 10 * x   

# Plotting
plt.figure(figsize=(6, 6))

# Plot the line
plt.plot(x, y_line, label=r'$y = (x-2)**2 + 4$', color='green')
plt.xlabel('x -->')
plt.ylabel('y -->')
plt.grid(True)
plt.show()

# Define the range of x values
x = np.linspace(-10, 10, 400)

# Define the functions with a larger coefficient for the linear term
y_parabola = x**2         
y_line = (x-2)**2 + 4                

# Plot the parabola
plt.plot(x, y_parabola, label=r'$y = x^2$', color='blue')

# Plot the line
plt.plot(x, y_line, label=r'$y = (x-2)**2 + 4$', color='green')

# Plot the combined function
#plt.plot(x, y_combined, label=r'$y = 2x^2 + 10x$', color='red')

# Adding labels and title
plt.xlabel('x -->')
plt.ylabel('y -->')


# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

## Visualization for different lambda values in Ridge regression
- Notice that in the cost function curve, the point with slope = 0 ( i.e. minima ), i.e. beta_1 value is reducing as lambda is increasing

```python
''' Generated by chat GPT using the following prompt : 
Share code in python to : 
1. Generate 100 random datapoints for (x,y)
2. consider h(x) = beta_1*x as the linear function ... generate 100 random data points for beta_1
3. Use the cost function J(beta_1) = (1/2m)*sum(i=1 to m) [h(x_i) - y_i]^2 + lambda*(beta_1^2).... calculate the following : 
3.1 Case 1 : consider lambda = 0 - calculate J(beta_1) for each of beta_1 value generated earlier
3.2 Case 2 : generate 3 random lambda values .. and calculate J(beta_1) for all beta_1 values .. for each of the random lambda value
4. Plot the 4 J(beta_1) curves generated with J(beta_1) in y axis .. and beta_1 in x axis .... use scatter plot
'''

import numpy as np
import matplotlib.pyplot as plt

# 1. Generate 100 random data points for (x, y)
np.random.seed(42)  # For reproducibility
x = np.random.rand(100)  # Random x values between 0 and 1
y = 3 * x + np.random.randn(100) * 0.5  # y = 3x + some noise
plt.scatter(x,y)
plt.xlabel('x--->')
plt.ylabel('y--->')

plt.show()
# 2. Generate 2500 random data points for beta_1
beta_1_values = np.random.uniform(-1500, 4500, 2500)/1000  # Random beta_1 values between -1.5 and 4.5

# 3. Define the cost function J(beta_1)
def cost_function(beta_1, x, y, lambda_):
    m = len(y)
    h_x = beta_1 * x
    cost = (1 / (2 * m)) * np.sum((h_x - y) ** 2) + lambda_ * (beta_1 ** 2)
    return cost

# 3.1 Case 1: Calculate J(beta_1) for each beta_1 value with lambda = 0
lambda_0 = 0
J_beta_1_lambda_0 = [cost_function(beta_1, x, y, lambda_0) for beta_1 in beta_1_values]

# 3.2 Case 2: Generate 3 random lambda values and calculate J(beta_1) for each
lambda_values = np.random.uniform(0, 0.2, 3)  # 3 Random lambda values between 0 and 0.2
J_beta_1_curves = []

for lambda_ in lambda_values:
    J_curve = [cost_function(beta_1, x, y, lambda_) for beta_1 in beta_1_values]
    J_beta_1_curves.append(J_curve)

# 4. Plot the J(beta_1) curves using scatter plot
plt.figure(figsize=(12, 8))

# Scatter plot for lambda = 0
plt.scatter(beta_1_values, J_beta_1_lambda_0, label=f'lambda = {lambda_0}', color='black', marker='.')

# Scatter plots for each random lambda value
for i, lambda_ in enumerate(lambda_values):
    plt.scatter(beta_1_values, J_beta_1_curves[i], label=f'lambda = {lambda_:.2f}', marker='.')

plt.xlabel('beta_1')
plt.ylabel('J(beta_1)')
plt.title('Cost Function J(beta_1) for Different lambda Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate 100 random data points for (x, y)
np.random.seed(42)  # For reproducibility
x = np.random.rand(100)  # Random x values between 0 and 1
y = 3 * x + np.random.randn(100) * 0.5  # y = 3x + some noise
plt.scatter(x,y)
plt.xlabel('x--->')
plt.ylabel('y--->')

plt.show()
# 2. Generate 2500 random data points for beta_1
beta_1_values = np.random.uniform(-1500, 4500, 2500)/1000  # Random beta_1 values between -1.5 and 4.5

# 3. Define the cost function J(beta_1)
def cost_function(beta_1, x, y, lambda_):
    m = len(y)
    h_x = beta_1 * x
    cost = (1 / (2 * m)) * np.sum((h_x - y) ** 2) + lambda_ * np.abs(beta_1)
    return cost

# 3.1 Case 1: Calculate J(beta_1) for each beta_1 value with lambda = 0
lambda_0 = 0
J_beta_1_lambda_0 = [cost_function(beta_1, x, y, lambda_0) for beta_1 in beta_1_values]

# 3.2 Case 2: Generate 3 random lambda values and calculate J(beta_1) for each
lambda_values = np.random.uniform(0, 1, 5)  # 5 Random lambda values between 0 and 1
J_beta_1_curves = []

for lambda_ in lambda_values:
    J_curve = [cost_function(beta_1, x, y, lambda_) for beta_1 in beta_1_values]
    J_beta_1_curves.append(J_curve)

# 4. Plot the J(beta_1) curves using scatter plot
plt.figure(figsize=(12, 8))

# Scatter plot for lambda = 0
plt.scatter(beta_1_values, J_beta_1_lambda_0, label=f'lambda = {lambda_0}', color='black', marker='.')

# Scatter plots for each random lambda value
for i, lambda_ in enumerate(lambda_values):
    plt.scatter(beta_1_values, J_beta_1_curves[i], label=f'lambda = {lambda_:.2f}', marker='.')

plt.xlabel('beta_1')
plt.ylabel('J(beta_1)')
plt.title('Cost Function J(beta_1) for Different lambda Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold--Domain expertise
corr_features=correlation(X_train,0.85)
corr_features
```

```python
## drop features when correlation is more than 0.85 
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_train.shape,X_test.shape
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

# Training and prediction
lasso=Lasso() # Lasso needs a valid alpha value. Default alpha of 1 is used
lasso.fit(X_train_scaled,y_train)
y_pred=lasso.predict(X_test_scaled)

# Performance metrics
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)

plt.scatter(y_test,y_pred)

# Coefficients
print(lasso.coef_) # Notice that many coefficients become 0 ... possibly because the alpha value of 1 is very high
```

## Ridge Regression model

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

# Training and prediction
ridge=Ridge() # ridge() requires an alpha value to be specified... default alpha of 1 is used
ridge.fit(X_train_scaled,y_train)
y_pred=ridge.predict(X_test_scaled)

# Performance metrics
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)

plt.scatter(y_test,y_pred)

# Coefficients
print(ridge.coef_) # Notice that none of the coefficients become 0 ... but they become very small
```

```python
class GDRegressor:
    
    def __init__(self,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            # update all the coef and the intercept
            y_hat = np.dot(X_train,self.coef_) + self.intercept_
            #print("Shape of y_hat",y_hat.shape)
            intercept_der = -2 * np.mean(y_train - y_hat)
            self.intercept_ = self.intercept_ - (self.lr * intercept_der)
            
            coef_der = -2 * np.dot((y_train - y_hat),X_train)/X_train.shape[0]
            self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
gdr = GDRegressor(epochs=1000,learning_rate=0.5)
```

```python
gdr.fit(X_train,y_train)
```

```python
y_pred = gdr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

# Class 36.4.9 : Stochastic Gradient Descent - by CampusX
- https://youtu.be/V7KBAa_gh4c?list=PLKnIA16_Rmvbr7zKYQuBfsVkjoLcJgxHH

```python
class SGDRegressor:
    
    def __init__(self,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        
    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            for j in range(X_train.shape[0]):
                idx = np.random.randint(0,X_train.shape[0])
                
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
                
                intercept_der = -2 * (y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)
                
                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
import time

sgd = SGDRegressor(learning_rate=0.01,epochs=40)
start = time.time()
sgd.fit(X_train,y_train)
print("The time taken is",time.time() - start)
```

```python
y_pred = sgd.predict(X_test)
r2_score(y_test,y_pred)
```

### Leaning Schedule 
- To reduce fluctuations in Stochastic GD as the solution reaches minima, learning rate is reduced as # iterations increase.
- Following code can be added in the SGDRegressor class to implement if needed

```python
class SGDRegressor:
    
    def __init__(self,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.t0, self.t1 = 5, 50 # For learning schedule implementation. Values can be changed as per need

    def learning_rate(t):
        return self.t0/(t + self.t1)

    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            for j in range(X_train.shape[0]):
                idx = np.random.randint(0,X_train.shape[0])

                self.lr = learning_rate(i*X.shape[0] + j ) # Learning schedule
                
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
                
                intercept_der = -2 * (y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)
                
                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
import time

sgd = SGDRegressor(learning_rate=0.01,epochs=100)
start = time.time()
sgd.fit(X_train,y_train)
print("The time taken is",time.time() - start)
```

```python
y_pred = sgd.predict(X_test)
r2_score(y_test,y_pred)
```

# Class 36.4.10 : Mini-Batch Gradient Descent - by CampusX
- https://youtu.be/_scscQ4HVTY?list=PLKnIA16_Rmvbr7zKYQuBfsVkjoLcJgxHH

```python
import random

class MBGDRegressor:
    
    def __init__(self,batch_size,learning_rate=0.01,epochs=100):
        
        self.coef_ = None
        self.intercept_ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size
        
    def fit(self,X_train,y_train):
        # init your coefs
        self.intercept_ = 0
        self.coef_ = np.ones(X_train.shape[1])
        
        for i in range(self.epochs):
            
            for j in range(int(X_train.shape[0]/self.batch_size)):
                
                idx = random.sample(range(X_train.shape[0]),self.batch_size)
                
                y_hat = np.dot(X_train[idx],self.coef_) + self.intercept_
                #print("Shape of y_hat",y_hat.shape)
                intercept_der = -2 * np.mean(y_train[idx] - y_hat)
                self.intercept_ = self.intercept_ - (self.lr * intercept_der)

                coef_der = -2 * np.dot((y_train[idx] - y_hat),X_train[idx])
                self.coef_ = self.coef_ - (self.lr * coef_der)
        
        print(self.intercept_,self.coef_)
    
    def predict(self,X_test):
        return np.dot(X_test,self.coef_) + self.intercept_
```

```python
mbr = MBGDRegressor(batch_size=int(X_train.shape[0]/50),learning_rate=0.01,epochs=100)
mbr.fit(X_train,y_train)
```

```python
y_pred = mbr.predict(X_test)
r2_score(y_test,y_pred)
```

## Understanding of bias variance tradeoff for Ridge regression

```python
m = 100
X = 5 * np.random.rand(m, 1) - 2
y = 0.7 * X ** 2 - 2 * X + 3 + np.random.randn(m, 1)

plt.scatter(X, y)
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X.reshape(100,1),y.reshape(100),test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=15)

X_train = poly.fit_transform(X_train)
X_test = poly.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
from mlxtend.evaluate import bias_variance_decomp

alphas = np.linspace(0,30,100)

loss = []
bias = []
variance = []

for i in alphas:
    reg = Ridge(alpha=i)
    avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        reg, X_train, y_train, X_test, y_test, 
        loss='mse',
        random_seed=123)
    loss.append(avg_expected_loss)
    bias.append(avg_bias)
    variance.append(avg_var)
```

```python
plt.plot(alphas,loss,label='loss')
plt.plot(alphas,bias,label='Bias')
plt.plot(alphas,variance,label='Variance')
#plt.ylim(0,5)
plt.xlabel('Alpha')
plt.legend()
```
